*   Todos os amigos devem receber a mesma quantidade de figurinhas;

No entanto, a cabine do bondinho tem uma capacidade máxima de 50 pessoas por viagem. Precisamos criar um programa que verifique se todos os alunos e monitores podem ser transportados em apenas uma viagem.
No desafio, você irá desenvolver um programa no Google Colab capaz de resolver uma situação-problema usando input, int e estruturas condicionais(if, else).

A ideia é verificar se o número de pessoas (alunos + monitores) cabe em um bondinho com capacidade para até 50 pessoas.

```python
alunos = int(input("Digite o número de alunos: "))
monitores = int(input("Digite o número de monitores: "))
resposta_positiva = "Todos os alunos e monitores podem usar o bondinho."
resposta_negativa = "Não há espaço suficiente para todos os alunos e monitores."

if alunos + monitores <= 50:
  print(resposta_positiva)
else:
  print(resposta_negativa)
```

## Aula 07: Cálculo Iterativo

A proposta é criar uma tabuada automatizada usando Python. Isso permitirá que o usuário insira qualquer número e receba a tabuada completa de forma rápida e prática.

A ideia é criar um programa que, ao receber um número como entrada, mostre o resultado das multiplicações desse número com os números de 1 a 10:

```python
def tabuada():
  n = int(input("Digite um número: "))
  for i in range(1, 11):
    resultado = n * i
    print(f"{n} x {i} = {resultado}")
```

```python
tabuada()
```

### Aula 09: Desafio Final

Imagine que você está desenvolvendo um sistema simples de controle de estoque. O programa deverá apresentar...

Entrada:

* Lista de produtos e suas quantidades;
* Comandos de entrada e saída de produtos.

Saída esperada:

* Estoque atualizado após cada operação;
* Mensagens de erro se um produto sair com quantidade insuficiente.

```python
def gerenciar_estoque():
    # 1. Lista inicial de produtos e quantidades (Dicionário)
    estoque = {
        "Caneta": 50,
        "Caderno": 20,
        "Borracha": 15,
        "Lápis": 100,
        "Apontador": 30,
        "Mochila": 10,
        "Cola": 10
    }

    print("--- ESTOQUE INICIAL ---")
    print(estoque)
    print("-" * 30)

    # 2. Lista de Comandos (Simulando a entrada de dados)
    # Formato: (Ação, Produto, Quantidade)
    comandos = [
        ("entrada", "Caneta", 10),    # Adiciona 10 canetas
        ("saida", "Caderno", 5),      # Remove 5 cadernos
        ("saida", "Borracha", 20),    # Tenta remover mais do que tem (ERRO)
        ("entrada", "Lápis", 100),    # Produto novo
        ("saida", "Caneta", 500)      # Tenta remover muito (ERRO)
    ]

    # Processamento dos comandos
    print("\n--- PROCESSANDO COMANDOS ---")

    for acao, produto, qtd in comandos:
        print(f"Comando: {acao.upper()} de {qtd} un. de {produto}")

        # Normaliza se o produto existe ou cria com 0
        if produto not in estoque:
            estoque[produto] = 0

        if acao == "entrada":
            estoque[produto] += qtd
            print(f"✅ Sucesso! Novo saldo de {produto}: {estoque[produto]}")

        elif acao == "saida":
            if estoque[produto] >= qtd:
                estoque[produto] -= qtd
                print(f"✅ Sucesso! Novo saldo de {produto}: {estoque[produto]}")
            else:
                print(f"❌ ERRO: Estoque insuficiente para {produto}. "
                      f"Disponível: {estoque[produto]}, Solicitado: {qtd}")

        print("-" * 30)

    # 3. Saída Final
    print("\n--- ESTOQUE FINAL ATUALIZADO ---")
    for item, quantidade in estoque.items():
        print(f"{item}: {quantidade}")

# Executa a função
gerenciar_estoque()
```
